### 크롤링 시도
- 중국 이커머스
- 알테쉬
- 중국 직구

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 크롬 드라이버 경로 설정
driver_path = 'path_to_chromedriver'  # chromedriver 경로로 변경
service = Service(driver_path)
driver = webdriver.Chrome(service=service)

# 크롤링할 네이버 뉴스 URL
base_url = 'https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query=중국+이커머스'

# 페이지 수 설정
num_pages = 5  # 원하는 페이지 수만큼 설정

for page in range(1, num_pages + 1):
    url = base_url + str(page)
    driver.get(url)
    
    # 페이지 로딩을 기다림
    time.sleep(3)
    
    # 기사 리스트 가져오기
    articles = driver.find_elements(By.CSS_SELECTOR, "ul.type06_headline li dl")
    
    for article in articles:
        title_element = article.find_element(By.CSS_SELECTOR, "dt:not(.photo) a")
        title = title_element.text
        link = title_element.get_attribute("href")
        
        # 기사 제목과 링크 출력
        print(f"제목: {title}")
        print(f"링크: {link}")
        print("="*80)

driver.quit()
```

### 크롤링 결과 출력

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from webdriver_manager.chrome import ChromeDriverManager


def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    #chrome_options.add_argument('headless') #brower invisible 상태에서 제어 
    
    #chorme의 버전정보를 확인해서 제어할 수 있는 drive 설치 환경 조성 
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                              options=chrome_options)
    return driver

driver = set_chrome_driver()

# 크롤링할 네이버 뉴스 검색 결과 URL
base_url = "https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query=중국직구"

# 페이지 수 설정
num_pages = 10  # 원하는 페이지 수만큼 설정

for page in range(1, num_pages + 1):
    url = base_url + "&start=" + str((page - 1) * 10 + 1)
    driver.get(url)
    
    # 페이지 로딩을 기다림
    time.sleep(3)
    
    # 기사 리스트 가져오기
    articles = driver.find_elements(By.CSS_SELECTOR, "ul.list_news li div.news_wrap.api_ani_send")
    
    for article in articles:
        title_element = article.find_element(By.CSS_SELECTOR, "a.news_tit")
        title = title_element.text
        link = title_element.get_attribute("href")
        
        # 기사 제목과 링크 출력
        print(f"제목: {title}")
        print(f"링크: {link}")
        print("="*80)

driver.quit()
```
