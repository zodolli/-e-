```python
import plotly.graph_objects as go

# 예제 데이터프레임 (df)
import pandas as pd


# 3자리 콤마를 넣은 문자열 생성 함수
def format_number(num):
    return f'{num:,}'

# 수입 건수 텍스트에 콤마 추가
df['수입 건수 (콤마)'] = df['수입 건수'].apply(format_number)

# 그래프 객체 생성
fig = go.Figure()

# 수입 건수 (막대 그래프)
fig.add_trace(go.Bar(x=df['기간'], y=df['수입 건수'], name='수입 건수',
                     text=df['수입 건수 (콤마)'], textposition='outside',
                     hoverinfo='x+y+text', hoverlabel=dict(bgcolor='white', font_size=12),
                     yaxis='y'))

# 수입 금액 (선 그래프)
fig.add_trace(go.Scatter(x=df['기간'], y=df['수입 금액'], mode='lines+markers', name='수입 금액',
                         text=df['수입 금액'], textposition='top center', line=dict(color='red'),
                         hoverinfo='x+y+text', hoverlabel=dict(bgcolor='white', font_size=12),
                         yaxis='y2'))

# 차트 레이아웃 설정
fig.update_layout(
    title='연도별 중국 수입 통계:건수 및 금액',
    xaxis_title='기간',
    yaxis=dict(
        title='수입 건수',
        side='left'
    ),
    yaxis2=dict(
        title='수입 금액',
        side='right',
        overlaying='y'
    ),
    hovermode='closest',
    width=1000,
    template='plotly_white',
    annotations=[dict(text='*단위: 건, 천 달러', x=0, y=-0.17, xref="paper", yref="paper", showarrow=False)],
    plot_bgcolor='white'
)

# 차트 표시
fig.show()
```
