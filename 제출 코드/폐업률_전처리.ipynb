```python
import pandas as pd

# 청크 단위로 파일 읽기 및 문제 있는 행 식별하기
def identify_problematic_rows_chunked(file_path, encoding='cp949', delimiter=',', expected_columns=33, chunksize=10000):
    problematic_rows = []
    chunk_iter = pd.read_csv(file_path, encoding=encoding, delimiter=delimiter, chunksize=chunksize, on_bad_lines='skip')
    
    for chunk_idx, chunk in enumerate(chunk_iter):
        for row_idx, row in chunk.iterrows():
            if len(row) != expected_columns:
                problematic_rows.append((chunk_idx * chunksize + row_idx + 1, row))

    return problematic_rows

# 문제 있는 행 가져오기
file_path = './data/fulldata_08_26_04_P_통신판매업.csv'
problematic_rows = identify_problematic_rows_chunked(file_path)

# 문제 있는 행 표시
for row in problematic_rows:
    print(f"Line {row[0]}: {row[1]}")

# 문제가 있는 경우 몇 줄만 미리보기로 확인
if problematic_rows:
    print(f"총 {len(problematic_rows)}개의 문제 있는 행이 발견되었습니다.")
    print("문제 있는 행 몇 가지 예시:")
    for row in problematic_rows[:5]:  # 첫 5개의 문제 있는 행만 출력
        print(f"Line {row[0]}: {row[1]}")
else:
    print("문제 있는 행이 발견되지 않았습니다.")
```
