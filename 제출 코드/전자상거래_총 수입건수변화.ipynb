```python
# 데이터 프레임 생성
data_2 = {
    '년도': [2023, 2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015],
    '수입(건수)': [34484637, 31608361, 29825928, 22855568, 17455201, 13107143, 10403481, 7630883, 6785600]
}
```

```python
df_2=pd.DataFrame(data_2)
df_2
```

```python
# 각 점에 표시될 텍스트 생성 (쉼표가 있는 형식으로)
text_values_2 = df_2['수입(건수)'].apply(lambda x: f'{x:,}')
```

```python
#그래프 생성
fig=go.Figure()

fig.add_trace(go.Scatter(x=df_2['년도'], y=df_2['수입(건수)'],
                         mode='lines+markers+text',
                         text=text_values_2,  #각 점에 해당하는 수출(건수)를 표시
                         textposition='top center',
                         textfont=dict(size=9)))  #텍스트 위치 설정

fig.update_layout(
    title='연도별 총 수입 건수 변화',
    xaxis_title='연도',
    yaxis_title='총 수입 건수',
    template='plotly_white',#템플릿 설정
    yaxis=dict(
        tickformat='~s')
)


fig.show()
```
